```python
import pandas as pd
data_west=pd.read_csv('WesternDiet_sample_set.csv')
```

```python
data_west
```

```python
data_west=data_west.rename(columns={'Unnamed: 0': 'id'})
```

```python
#check if there are missing values in the dataset
#if there are, print yes and further cleaning is required
import numpy as np
for column in data_west:
    for row in data_west[column]:
        if row=='n.a.':
            print('yes')
```

```python
#choose which column has the value type 'string'
type_string=['id']
error_column=[]
error_row=[]
for column in data_west:
    for row in range (len(data_west[column])):
        if column not in type_string:
            #change all the values in this column to type 'int'
            #if error occurs, that is the missing value and take note its position
            try:
                data_west.loc[row,column]=int(data_west.loc[row,column])
            except (TypeError,ValueError):
                error_column.append(column)
                error_row.append(row)
#for each missing value, replace it with the median of that column wihtout the missing value
for i in range (len(error_column)):
    row=error_row[i]
    column=error_column[i]
    data_west.loc[row, column]=np.median(data_west.loc[list(range(0,row))+list(range(row+1,len(data_west[column]))),column])

```

```python
#another check on the availability of the missing value
for column in data_west:
    for row in data_west[column]:
        if row=='n.a.':
            print('yes')
```

```python
for i in range (len(error_column)):
    row=error_row[i]
    column=error_column[i]
    print(data_west.loc[row,column])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#for this dataset, batch is always 1
n=len(data_west['id'])
data_west['batch']=pd.Series([1]*n)
#select all the QCs as the data to build the linear regression model
QC1s=data_west.loc[data_west['id']=='QC1']
QC2s=data_west.loc[data_west['id']=='QC2']
QCs=pd.concat([QC1s,QC2s],ignore_index=True)
#take batch column as the x axis and each column as y axis to build model
x=QCs.loc[:,'batch'].values.reshape(-1,1)
for column in data_west:
    if column !='id' and column!='batch':
        y=QCs.loc[:,column].values.reshape(-1,1)
        reg = LinearRegression()
        reg.fit(x,y)
        for row in range (len(data_west[column])):
            #use predict function to predict each row value in that column and take it as x_hat
            #x_corrected = x_uncorrected - x_hat + x_mean
            column_hat=reg.predict(data_west.loc[row,'batch'].reshape(-1,1))
            column_hat=column_hat[0][0]
            data_west.loc[row,column]=data_west.loc[row, column]-column_hat+np.mean(data_west[column])
```

```python
data_west.to_csv('CorrectedWestDiet.csv')
```

```python
type_string=['ID']
for column in data_japan:
    for row in range (len(data_japan[column])):
        if column not in type_string:
            data_japan.loc[row,column]=int(data_japan.loc[row,column])
            if column != 'batch':
                if data_japan.loc[row,column] <LOD[column]:
                    data_japan.loc[row,column]=LOD[column]
```

```python
dummies=pd.get_dummies(data_japan['batch'])
data_japan=data_japan.join(dummies.iloc[:,0:5], lsuffix='_caller', rsuffix='_other')
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
QC1s=data_japan.loc[data_japan['ID']=='QC1']
QC2s=data_japan.loc[data_japan['ID']=='QC2']
QCs=pd.concat([QC1s,QC2s],ignore_index=True)
x=QC2s.loc[:,[1,2,3,4,5]].values
for column in data_japan:
    if column !='ID' and column!='batch' and column not in [1,2,3,4,5]:
        y=QC2s.loc[:,column].values.reshape(-1,1)
        reg = LinearRegression()
        reg.fit(x,y)
        for row in range (len(data_japan[column])):
            column_hat=reg.predict(data_japan.loc[row,[1,2,3,4,5]].values.reshape(1,-1))
            column_hat=column_hat[0][0]
            data_japan.loc[row,column]=data_japan.loc[row, column]-column_hat+np.mean(data_japan[column])
```
