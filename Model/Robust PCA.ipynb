```python
LOD={'C14:0':23953,
'C15:0':4879,
'C16:0':21972,
'trans C16:1n-7':708,
'C16:1n-7':3174,
'C17:0':6404,
'C18:0':13327,
'trans C18:1n-9':422,
'trans C18:1n-7':471,
'C18:1n-9':2179,
'C18:1n-7':642,
'trans C18:2n-6':1154,
'C18:2 n-6, LA':2588,
'C18:3 n-6, GLA':1741,
'C18:3 n-3, ALA':5608,
'C20:0':8926,
'C20:1n-9':545,
'C20:2n-6':735,
'C20:3n-6':1476,
'C20:4n-6,AA':1105,
'C22:0':3507,
'C20:5n-3,EPA':1044,
'C22:4n-6':3831,
'C24:0':2636,
'C24:1n-9':540,
'C22:5n-3,DPA':904,
'C22:6n-3,DHA':901
}
```

```python
type_string=['ID']
for column in data_japan:
    for row in range (len(data_japan[column])):
        if column not in type_string:
            data_japan.loc[row,column]=int(data_japan.loc[row,column])
            if column != 'batch':
                if data_japan.loc[row,column] <LOD[column]:
                    data_japan.loc[row,column]=LOD[column]
```

```python
column_no_standardlized=['ID','batch']
for column in data_japan:
    if column not in column_no_standardlized:
        data_japan[column]=(data_japan[column]-data_japan[column].mean())/(data_japan[column].std())
```

```python
data_japan
```

```python
data_rpca=data_japan.drop(columns=['batch','ID'])
```

```python
from hyperspy.signals import Signal1D
```

```python
s = Signal1D(data_rpca)
```

```python
s.decomposition(algorithm="RPCA", output_dimension=2)
```

```python
result_rpca=pd.DataFrame(s.learning_results.loadings, columns=['RPC1','RPC2'])
```
